## Guassian Transformation

```python
df = pd.read_csv("D:\\ML\\data\\titanic.csv", usecols=['Age','Fare','Survived'])
```

```python
df['Age']= df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
import scipy.stats as stat
import pylab
```

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'Age')
```

## Logarithmatic Transformation

```python
import numpy as np
df['Age_log']= np.log(df['Age'])
plot_data(df,'Age_log')
```

## Reciprocal transaformation

```python
df['Age_reci'] = 1/df.Age
```

```python
plot_data(df,'Age_reci')
```

## Square Root Transformation

```python
df['Age_squr'] = df.Age**(1/2)
plot_data(df,'Age_squr')
```

## Exponencial Transformation

```python
df['Age_Expon'] = df.Age*(1/1.2)
plot_data(df,'Age_Expon')
```

## BoxCox transformation

```python
df['Age_boxcox'],parameters = stat.boxcox(df['Age'])
```

```python
print(parameters)
```

```python
plot_data(df,'Age_boxcox')
```

```python
plot_data(df,'Fare')
```

```python
#### Fare
df['Fare_log']=np.log1p(df['Fare'])
plot_data(df,'Fare_log')
```

```python
df['Fare_Boxcox'],parameters=stat.boxcox(df['Fare']+1)
plot_data(df,'Fare_Boxcox')
```
